### Importing Data

```python
### pyodbc connection
conn = pyodbc.connect('Driver={SQL Server};'
                      'Server=BMA-SQL14-SB09\SB09;'
                      'Database=SBDW;'
                      'Trusted_Connection=yes;'
                      )

query = """
        SELECT top 30000 
        b.CustomerKey,
        e.EventKey,
        e.EventName,
        e.EventDeadline,
        b.BetKey,
        bs.BetSelectionKey,
        b.CouponArrivedDate,
        m.MarketKey,
        m.MarketName,
        sc.SubCategoryName as LeagueName, 
        c.CategoryName as SportName,
        z.SubCategoryGroupName as ZoneName
        
        FROM [DW].[MarketSelection] ms WITH(NOLOCK)
        INNER JOIN DW.BetSelection bs WITH(NOLOCK)
        	ON ms.MarketSelectionKey = bs.MarketSelectionKey
        INNER JOIN DW.Market m with(NOLOCK) 
        	ON ms.MarketKey = m.MarketKey
        INNER JOIN DW.Event e WITH(NOLOCK)
        	ON e.EventKey = m.EventKey
        INNER JOIN DW.SubCategory sc WITH(NOLOCK)
        	ON e.SubCategoryKey = sc.SubCategoryKey
        INNER JOIN DW.Category c WITH(NOLOCK)
        	ON c.CategoryKey = sc.CategoryKey
        INNER JOIN DW.Bet b WITH(NOLOCK)
        	ON bs.BetKey = b.BetKey
        inner join DW.SubCategoryGroup z WITH(NOLOCK)
        	on sc.SubCategoryGroupKey = z.SubCategoryGroupKey
        where CouponArrivedDate  > CAST(DATEADD(d, -10, GetDate()) as date)
        and EventDeadline  > CAST(DATEADD(d, -10, GetDate()) as date)
        and EventDeadline  <= CAST(DATEADD(d, 3, GetDate()) as date)
        and e.EventTypeKey = 2
        """

df = pd.read_sql(query, conn)
```
